## Recursion
- pyhton allow limited recursion and if the number exceeds then it will through recursionerror, inorder to prevent infinite recursion so that we don't have memory overuse issue.

```python
## infinite recursion example
def greet(i=0):
    i+=1
    print("Hello, OM!",i)
    greet(i)  # This will cause infinite recursion

greet()  # Call the function to see the recursion in action
```

```python
def func(count=0):
    if count == 4:
        return
    print("Omi!!")
    count += 1
    func(count)
func()
```
