```python
## Read & Write files...

# Read whole file

with open('example.txt', 'r') as file :
    content = file.read()
    print(content)
```

```python
## read file line by line

#note the space between both of the outputs.

with open('example.txt', 'r') as file:
    for line in file:
        print(line)
```

```python
## Writing a file(overwriting)

with open('example.txt', 'w') as file:
    file.write("Created by Omii\n")
    file.write("This is the created line")
```

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line)
```

```python
## Writing a file(Without Overwriting)

with open('example.txt', 'a') as file:
    file.write("Unfortunately the main file was deleted in previous operation\n")
    file.write("This is the append file.")
```

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line)
```

```python
## Writing a list of lines to a file

lines = ['\n new appended line - 1', '\n new appended line - 2', '\n new appended line - 3', '\n These are lines that were appended at once.']

with open('example.txt', 'a') as file:
    file.writelines(lines)

with open('example.txt', 'r') as file:
    for line in file:
        print(line)
```

```python
## Handling Binary File

# writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)

# reading binary file
with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
## Checking if path is a file or directory

import os 

path = 'example.txt'

if os.path.exists(path):
    print(f"The path {path} is a file.")
elif os.path.isdir(path):
    print(f"The path {path} is a directory")

else:
    print(f"The path {path} neither file nor directory.")
```
